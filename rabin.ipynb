# BASED ON TIME

```python
mins=list(map(str, range(1,121)))#Creates a list of strings representing minutes from 1 to 120 using the range() function and the map() function to convert each integer to a string
mins.insert(45,"1stE")
mins.insert(91,"2ndE")
mins.insert(122,"extra")
data.loc[data.Minute.str[:3]=='45+', 'Minute'] = '1stE'
data.loc[data.Minute.str[:3]=='90+', 'Minute'] = '2ndE'
data.loc[data.Minute.str[:4]=='120+', 'Minute'] = 'Extra'
mins1=mins[:23]
mins2=mins[23:46]
mins3=mins[46:69]
mins4=mins[69:92]
mins5=mins[92:]
```

```python
sns.set(rc={'figure.figsize':(20,7)})
plt.xticks(fontsize=20)
p=sns.countplot(x='Minute',data=data,order=mins1)
p.axes.set_title("Goals per Game Minute (1-23)",fontsize=35)
```

```python
p=sns.countplot(x='Minute',data=data,order=mins2)
p.axes.set_title("Goals per Game Minute (24-First Half Extra time)",fontsize=30)
```

```python
p=sns.countplot(x='Minute',data=data,order=mins3)
p.axes.set_title("Goals per Game Minute (46-68)",fontsize=30)
```

```python
p=sns.countplot(x='Minute',data=data,order=mins4)
p.axes.set_title("Goals per Game Minute 69- 2nd Half Extra time)",fontsize=30)
```

```python
p=sns.countplot(x='Minute',data=data,order=mins5)
p.axes.set_title("Goals per Game Minute (91-Extra times)",fontsize=30)
```

# Goals per opponent

```python
mins=list(map(str, data.Opponent.value_counts().sort_values(ascending=False).index))
for min in data['Opponent']:
    if min not in mins:
        mins.append(min)
mins1=mins[:int(len(mins)/5)]
mins2=mins[int(len(mins)/5):int(2*len(mins)/5)]
mins3=mins[2*int(len(mins)/5):int(3*len(mins)/5)]
mins4=mins[3*int(len(mins)/5):int(4*len(mins)/5)]
mins5=mins[int(4*len(mins)/5):]
```

```python
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(fontsize=15,rotation='vertical')
p=sns.countplot(x='Opponent',data=data,order=mins1)
p.axes.set_title("Goals per Opponents",fontsize=30)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(fontsize=15,rotation='vertical')
p=sns.countplot(x='Opponent',data=data,order=mins2)
p.axes.set_title("Goals per Opponents",fontsize=30)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(fontsize=15,rotation='vertical')
p=sns.countplot(x='Opponent',data=data,order=mins3)
p.axes.set_title("Goals per Opponents",fontsize=30)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(fontsize=15,rotation='vertical')
p=sns.countplot(x='Opponent',data=data,order=mins4)
p.axes.set_title("Goals per Opponents",fontsize=30)
```
